# Smart Contract paper 1 - BlockChain Samrt contract.txt

```python
file = open('BlockChain Samrt contract.txt',encoding='utf-8')
data = file.read()
file.close()
```

```python
def random_color_func(word=None, font_size=None, position=None, orientation=None, font_path=None, random_state=None):
    h = int(360.0 * 45.0 / 255.0)
    s = int(100.0 * 255.0 / 255.0)
    l = int(100.0 * float(random_state.randint(60, 120)) / 255.0)

    return "hsl({}, {}%, {}%)".format(h, s, l)
```

```python
wordcloud = WordCloud(stopwords = STOPWORDS,
                            background_color = 'white',
                            width = 2400,
                            height = 2000,
                            color_func = random_color_func
                            ).generate(data)

figure(figsize=(10, 8), dpi=80)
plt.imshow(wordcloud)
plt.axis('off')
plt.title("Word Cloud for 'BlockChain Samrt contract.txt' paper")
plt.show()
```

# SC paper 3 - BlockChain Samrt contract2.txt

```python
file = open('BlockChain Samrt contract2.txt',encoding='utf-8')
data = file.read()
file.close()

wordcloud = WordCloud(stopwords = STOPWORDS,
                            background_color = 'white',
                            width = 2400,
                            height = 2000,
                            color_func = random_color_func
                            ).generate(data)

figure(figsize=(10, 8), dpi=80)
plt.imshow(wordcloud)
plt.axis('off')
plt.title("Word Cloud for 'BlockChain Samrt contract2.txt' paper")
plt.show()
```

```python
data_tokens = word_tokenize(data)
data_without_sw = [word for word in data_tokens if not word in stopwords.words()]
        
data_final = ' '.join([str(n) for n in data_without_sw])

```

```python
words = data_final.split()
for i in words:
    if ((i == ',') or (i == '.') or(i == ')')or(i == '(')or(i == '[')or(i == ']')):
        words.remove(i)

for i in words:
    if ((i == ':') or (i == '"') or(i == '“')or(i == '”')or(i == 'The')or(i == '. ')):
        words.remove(i)

for i in words:
    if ((i == '2018') or (i == '2019') or(i == '2017')or(i == '4')or(i == '<')or(i == '>')):
        words.remove(i)

for i in words:
    if ((i == ',') or (i == ';') or(i == '2017')or(i == '[')or(i == '<')or(i == '>')):
        words.remove(i)
        
        
count = pd.value_counts(np.array(words))
d = count[:14]
df = d.to_frame()
df.columns = ['CNT']
df['index'] = df.index

d
```
